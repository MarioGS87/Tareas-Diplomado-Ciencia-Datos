```python

import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# Generar datos
np.random.seed(1995)
x1 = np.random.random(5) + 5
y1 = np.random.random(5) + 4

x2 = np.random.random(5) + 4
y2 = np.random.random(5) + 4

x3 = np.random.random(5) + 5
y3 = np.random.random(5) + 5

x4 = np.random.random(5) + 4
y4 = np.random.random(5) + 5

x = list(x1) + list(x2) + list(x3) + list(x4)
y = list(y1) + list(y2) + list(y3) + list(y4)
target = ['g'] * 5 + ['r'] * 15
data = pd.DataFrame({'x1': x, 'x2': y, 'y': target})

# Definir el rango de variables a considerar
num_variables = len(data.columns) - 1  # Excluyendo la variable objetivo

# Inicializar variables para almacenar los resultados
best_score = 0
best_num_features = 0
best_model = None

# Probar diferentes números de variables
for num_features in range(1, num_variables + 1):
    # Seleccionar las primeras 'num_features' variables como características
    features = data.iloc[:, :num_features]
    
    # Entrenar un modelo de árbol de decisión utilizando validación cruzada
    model = DecisionTreeClassifier(random_state=42)
    scores = cross_val_score(model, features, data['y'], cv=5)  # 5-fold cross-validation
    
    # Calcular la puntuación media de validación cruzada
    avg_score = scores.mean()
    
    # Actualizar el mejor modelo si se encuentra un nuevo mejor
    if avg_score > best_score:
        best_score = avg_score
        best_num_features = num_features
        best_model = model

print(f"Mejor número de variables: {best_num_features}")
print(f"Puntuación promedio de validación cruzada: {best_score:.4f}")
print("Variables seleccionadas:")
print(data.columns[:best_num_features])
```

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Datos
np.random.seed(1995)
x1 = np.random.random(5) + 5
y1 = np.random.random(5) + 4

x2 = np.random.random(5) + 4
y2 = np.random.random(5) + 4

x3 = np.random.random(5) + 5
y3 = np.random.random(5) + 5

x4 = np.random.random(5) + 4
y4 = np.random.random(5) + 5

x = list(x1) + list(x2) + list(x3) + list(x4)
y = list(y1) + list(y2) + list(y3) + list(y4)
target = ['g'] * 5 + ['r'] * 15
data = pd.DataFrame({'x1': x, 'x2': y, 'y': target})

# Definir características (variables independientes) y variable objetivo
X = data[['x1', 'x2']]
y = data['y']

# Definir modelo de árbol de decisión
model = DecisionTreeClassifier()

# Definir parámetros para búsqueda de hiperparámetros
param_grid = {
    'max_depth': [2, 3, 4, 5],
    'min_samples_split': [2, 3, 4, 5],
    'min_samples_leaf': [1, 2, 3, 4, 5]
}

# Búsqueda de hiperparámetros utilizando validación cruzada
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X, y)

# Mostrar mejores parámetros encontrados
print("Mejores parámetros encontrados:")
print(grid_search.best_params_)
```

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

# Generar datos
np.random.seed(1995)
x1 = np.random.random(5) + 5
y1 = np.random.random(5) + 4

x2 = np.random.random(5) + 4
y2 = np.random.random(5) + 4

x3 = np.random.random(5) + 5
y3 = np.random.random(5) + 5

x4 = np.random.random(5) + 4
y4 = np.random.random(5) + 5

x = list(x1) + list(x2) + list(x3) + list(x4)
y = list(y1) + list(y2) + list(y3) + list(y4)
target = ['g'] * 5 + ['r'] * 15
data = pd.DataFrame({'x1': x, 'x2': y, 'y': target})

# Definir función para encontrar los puntos de corte
def find_split_points(data, feature):
    # Ordenar los valores únicos del feature
    unique_values = sorted(data[feature].unique())
    split_points = []
    for i in range(len(unique_values) - 1):
        split_point = (unique_values[i] + unique_values[i+1]) / 2
        split_points.append(split_point)
    return split_points

# Encontrar los puntos de corte para cada característica
split_points = {}
for feature in data.columns[:-1]:
    split_points[feature] = find_split_points(data, feature)

# Mostrar los puntos de corte encontrados
for feature, points in split_points.items():
    print(f"Puntos de corte para '{feature}': {points}")
```

```python
import numpy as np
import pandas as pd

# Generar datos
np.random.seed(1995)
x1 = np.random.random(5) + 5
y1 = np.random.random(5) + 4

x2 = np.random.random(5) + 4
y2 = np.random.random(5) + 4

x3 = np.random.random(5) + 5
y3 = np.random.random(5) + 5

x4 = np.random.random(5) + 4
y4 = np.random.random(5) + 5

x = list(x1) + list(x2) + list(x3) + list(x4)
y = list(y1) + list(y2) + list(y3) + list(y4)
target = ['g'] * 5 + ['r'] * 15
data = pd.DataFrame({'x1': x, 'x2': y, 'y': target})

# Función para calcular la entropía
def entropy(labels):
    n_labels = len(labels)
    if n_labels <= 1:
        return 0
    value, counts = np.unique(labels, return_counts=True)
    probs = counts / n_labels
    n_classes = np.count_nonzero(probs)
    if n_classes <= 1:
        return 0
    ent = 0.
    for prob in probs:
        ent -= prob * np.log(prob)
    return ent

# Función para encontrar el punto de división óptimo basado en la entropía
def find_split_point(data, feature):
    sorted_data = data.sort_values(feature)
    labels = sorted_data['y'].values
    best_entropy = float('inf')
    best_split_point = None
    for i in range(1, len(labels)):
        left_entropy = entropy(labels[:i])
        right_entropy = entropy(labels[i:])
        total_entropy = (i / len(labels)) * left_entropy + ((len(labels) - i) / len(labels)) * right_entropy
        if total_entropy < best_entropy:
            best_entropy = total_entropy
            best_split_point = i
    return best_split_point

# Encontrar el punto de división óptimo para cada característica
split_points = {}
for feature in ['x1', 'x2']:
    split_point = find_split_point(data, feature)
    split_points[feature] = split_point

print("Puntos de división óptimos para cada característica:")
print(split_points)
```
