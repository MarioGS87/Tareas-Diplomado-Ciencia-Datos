```python
# Ejercicio en clase hecho por Mario 04/05/24 9:20am

import numpy as np

def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((np.array(point1) - np.array(point2)) ** 2))

def min_distance(cluster1, cluster2):
    min_dist = float('inf')
    for p1 in cluster1:
        for p2 in cluster2:
            dist = euclidean_distance(p1, p2)
            if dist < min_dist:
                min_dist = dist
    return min_dist

def max_distance(cluster1, cluster2):
    max_dist = 0
    for p1 in cluster1:
        for p2 in cluster2:
            dist = euclidean_distance(p1, p2)
            if dist > max_dist:
                max_dist = dist
    return max_dist

def average_distance(cluster1, cluster2):
    total_dist = 0
    count = 0
    for p1 in cluster1:
        for p2 in cluster2:
            total_dist += euclidean_distance(p1, p2)
            count += 1
    return total_dist / count

def centroid_distance(cluster1, cluster2):
    centroid1 = np.mean(cluster1, axis=0)
    centroid2 = np.mean(cluster2, axis=0)
    return euclidean_distance(centroid1, centroid2)


def hierarchical_clustering(points, distance_metric):
    clusters = [[point] for point in points]
    while len(clusters) > 1:
        min_dist = float('inf')
        closest_clusters = None
        for i in range(len(clusters)):
            for j in range(i+1, len(clusters)):
                dist = distance_metric(clusters[i], clusters[j])
                if dist < min_dist:
                    min_dist = dist
                    closest_clusters = (i, j)
        
        # Merge closest clusters
        i, j = closest_clusters
        clusters[i].extend(clusters[j])
        del clusters[j]

        print(f"Merging clusters {i} and {j} with distance {min_dist}")

    return clusters

points = [
    [5,3], [10,15], [15,12], [24,10], [30,30],
    [85,70], [71,80], [60,78], [70,55], [80,91]
]

# Try each distance metric
distance_metrics = {
    "Min Distance": min_distance,
    "Max Distance": max_distance,
    "Average Distance": average_distance,
    "Centroid Distance": centroid_distance
}

for name, metric in distance_metrics.items():
    print(f"Using {name}:")
    hierarchical_clustering(points, metric)
    print()
```
