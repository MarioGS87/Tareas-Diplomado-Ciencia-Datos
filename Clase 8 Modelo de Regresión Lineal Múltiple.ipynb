```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Leer los datos desde el archivo CSV
data = pd.read_csv('data_2d.csv', names = ['X1', 'X2', 'y'])

# Separar las variables independientes (x1, x2) de la variable dependiente (y)
X = data[['X1', 'X2']]
y = data['y']

# Crear un modelo de regresión lineal múltiple
modelo = LinearRegression()

# Entrenar el modelo con los datos
modelo.fit(X, y)

# Coeficientes de la regresión (intercepto y pendientes)
intercepto = modelo.intercept_
coeficientes = modelo.coef_

# Crear una figura tridimensional
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos de datos
ax.scatter(data['X1'], data['X2'], data['y'], color='blue', label='Datos')

# Crear una malla de puntos en el espacio tridimensional
x1_range = data['X1'].values
x2_range = data['X2'].values
x1_range, x2_range = np.meshgrid(x1_range, x2_range)
y_pred = modelo.predict(pd.DataFrame({'X1': x1_range.ravel(), 'X2': x2_range.ravel()}))
y_pred = y_pred.reshape(x1_range.shape)

# Graficar el plano de regresión lineal
ax.plot_surface(x1_range, x2_range, y_pred, color='green', alpha=0.5, label='Plano de regresión')

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
ax.set_title('Regresión Lineal Múltiple')
```
